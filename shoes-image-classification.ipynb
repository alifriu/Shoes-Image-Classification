### Menampilkan Resolusi Gambar

```python
def print_images_resolution(directory):
    total_images = 0
    unique_sizes = {}

    for subdir in os.listdir(directory):
        subdir_path = os.path.join(directory, subdir)
        image_files = os.listdir(subdir_path)
        num_images = len(image_files)
        total_images += num_images

        for img_file in image_files:
            img_path = os.path.join(subdir_path, img_file)
            with Image.open(img_path) as img:
                size = img.size
                unique_sizes[size] = unique_sizes.get(size, 0) + 1

    print(f"Directory: {directory}")
    print(f"Total Images: {total_images}")
    print("Image Resolutions (Count):")
    for size, count in unique_sizes.items():
        print(f"- {size} ({count})")

# Example usage
print_images_resolution('/kaggle/working/dataset')
```

```python
# Panggil variabel mypath yang menampung folder dataset gambar
mypath= '/kaggle/working/dataset/'

file_name = []
labels = []
full_path = []
for path, subdirs, files in os.walk(mypath):
    for name in files:
        full_path.append(os.path.join(path, name))
        labels.append(path.split('/')[-1])
        file_name.append(name)


# Memasukan variabel yang sudah dikumpulkan pada looping di atas menjadi sebuah dataframe agar rapih
df = pd.DataFrame({"path":full_path,'file_name':file_name,"labels":labels})
# Melihat jumlah data gambar pada masing-masing label
df.groupby(['labels']).size()
```

#### Image Data Generator

```python
# Define training and test directories
TRAIN_DIR = "/kaggle/working/Dataset-Final/train"
TEST_DIR = "/kaggle/working/Dataset-Final/test"

train_balletFlat = os.path.join(TRAIN_DIR + '/Ballet Flat')
train_boat = os.path.join(TRAIN_DIR + '/Boat')
train_brogue = os.path.join(TRAIN_DIR + '/Brogue')
train_clog = os.path.join(TRAIN_DIR + '/Clog')
train_sneaker = os.path.join(TRAIN_DIR + '/Sneaker')

test_balletFlat = os.path.join(TEST_DIR + '/Ballet Flat')
test_boat = os.path.join(TEST_DIR + '/Boat')
test_brogue = os.path.join(TEST_DIR + '/Brogue')
test_clog = os.path.join(TEST_DIR + '/Clog')
test_sneaker = os.path.join(TEST_DIR + '/Sneaker')

print("Total number of ballet flat images in training set: ",len(os.listdir(train_balletFlat)))
print("Total number of boat images in training set: ",len(os.listdir(train_boat)))
print("Total number of brogue images in training set: ",len(os.listdir(train_brogue)))
print("Total number of clog images in training set: ",len(os.listdir(train_clog)))
print("Total number of sneaker images in training set: ",len(os.listdir(train_sneaker)))
print("\n")
print("Total number of ballet flat images in test set: ",len(os.listdir(test_balletFlat)))
print("Total number of boat images in test set: ",len(os.listdir(test_boat)))
print("Total number of brogue images in test set: ",len(os.listdir(test_brogue)))
print("Total number of clog images in test set: ",len(os.listdir(test_clog)))
print("Total number of sneaker images in test set: ",len(os.listdir(test_sneaker)))
```

```python
# Create an ImageDataGenerator object normalizing the images
datagen = ImageDataGenerator(rescale=1/255.,
                             rotation_range=30,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             shear_range=0.2,
                             zoom_range=0.2,
                             horizontal_flip=True,
                             vertical_flip=True,
                             )
val_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = datagen.flow_from_directory(TRAIN_DIR,
                                              batch_size=32,
                                              target_size=(224,224),
                                              color_mode="rgb",
                                              class_mode='categorical',
                                              shuffle=True)

validation_generator = val_datagen.flow_from_directory(TEST_DIR,
                                                   batch_size=32,
                                                   target_size=(224,224),
                                                   color_mode="rgb",
                                                   class_mode='categorical',
                                                   shuffle=False)
```

## Modelling

```python
# Membentuk Model CNN dengan Sequential()
model = Sequential()

# Menggunakan ResNet50V2 sebagai base model
base_model = ResNet50V2(weights="imagenet", include_top=False, input_tensor=Input(shape=(224, 224, 3)))

# Membekukan semua layer pada base model
for layer in base_model.layers:
    layer.trainable = False

# Menambahkan base model ke dalam model Sequential
model.add(base_model)

# Layer tambahan pertama
model.add(Conv2D(256, (3, 3), activation='relu', padding='same'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(BatchNormalization(momentum=0.9))
model.add(Dropout(0.2))

# Layer tambahan kedua
model.add(GlobalAveragePooling2D())

# Layer fully connected pertama
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization(momentum=0.9))
model.add(Dropout(0.2))

# Layer fully connected kedua
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization(momentum=0.9))

# Layer output
model.add(Dense(5, activation='softmax'))

# Compile model dengan optimizer Adam dan metrik akurasi
model.compile(loss="categorical_crossentropy",
              optimizer=optimizers.Adam(learning_rate=0.001),
              metrics=["accuracy"])

# Menampilkan bentuk model yang telah dibuat
model.summary()
```

```python
# Mengimplementasikan callback dengan batas bawah akurasi 97%
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs = {}):
        if (logs.get("accuracy") > 0.85) and (logs.get("val_accuracy") > 0.85):
            print("\nModel selesai dilatih karena akurasi telah lebih dari 85%!")
            self.model.stop_training = True

callbacks = myCallback()

reduce_lr = ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=3, min_lr=1e-6)
```

```python
count_balletFlat= len(os.listdir(train_balletFlat))
count_boat = len(os.listdir(train_boat))
count_brogue = len(os.listdir(train_brogue))
count_clog = len(os.listdir(train_clog))
count_sneaker = len(os.listdir(train_sneaker))

weight_0 = (1 / count_balletFlat) * (count_balletFlat + count_boat + count_brogue + count_clog + count_sneaker) / 5.0
weight_1 = (1 / count_boat) * (count_balletFlat + count_boat + count_brogue + count_clog + count_sneaker) / 5.0
weight_2 = (1 / count_brogue) * (count_balletFlat + count_boat + count_brogue + count_clog + count_sneaker) / 5.0
weight_3 = (1 / count_clog) * (count_balletFlat + count_boat + count_brogue + count_clog + count_sneaker) / 5.0
weight_4 = (1 / count_sneaker) * (count_balletFlat + count_boat + count_brogue + count_clog + count_sneaker) / 5.0

class_weights = {0 : weight_0, 1 : weight_1, 2 : weight_2, 3 : weight_3, 4 : weight_4}

# Memulai timer
start_time = time.time()

history = model.fit(
    train_generator,
    epochs = 30,
    batch_size = 64,
    validation_data = validation_generator,
    class_weight = class_weights,
    callbacks = [callbacks,reduce_lr]
)

# Memberhentikan timer
end_time = time.time()

# Menghitung total waktu pelatihan untuk model
training_time = end_time - start_time
print(f"\nTotal waktu model dilatih: {training_time:.2f} detik")
```

## Evaluasi dan Visualisasi

```python
hasil_training = pd.DataFrame(history.history)

# Mendefinisikan variabel epoch yang merupakan jumlah epoch pada pelatihan
epoch = np.arange(len(hasil_training))

# Membentuk gambar berisi 2 sub gambar
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
fig.suptitle("Performa Model", fontsize = 18)

# Membentuk gambar yang menampilkan performa akurasi pada model
ax[0].plot(epoch, hasil_training.accuracy, label = "Akurasi Training", linewidth = 2, color = "black")
ax[0].plot(epoch, hasil_training.val_accuracy, label = "Akurasi Testing", linewidth = 2, color = "red")
ax[0].set_title("Akurasi Model", fontsize = 16)
ax[0].set_xlabel("Epochs", fontsize = 12)
ax[0].set_ylabel("Akurasi", fontsize = 12)
ax[0].legend()
ax[0].grid(True)

# Membentuk gambar yang menampilkan performa loss pada model
ax[1].plot(epoch, hasil_training.loss, label = "Training Loss", linewidth = 2, color = "black")
ax[1].plot(epoch, hasil_training.val_loss, label = "Testing Loss", linewidth = 2, color= "red")
ax[1].set_title("Loss Model", fontsize = 16)
ax[1].set_xlabel("Epochs", fontsize = 12)
ax[1].set_ylabel("Loss", fontsize = 12)
ax[1].legend()
ax[1].grid(True)

# Mengatur layout gambar
plt.tight_layout(rect = [0, 0, 1, 0.95])

# Menampilkan gambar
plt.show()
```
